```python
# Create an empty 3x3 Tic Tac Toe game board
def create_board():
    board = [[' ' for _ in range(3)] for _ in range(3)]
    return board

# Function to display the game board
def display_board(board):
    for row in board:
        print('|'.join(row))
        print('-' * 5)

# Function to check if a move is valid
def is_valid_move(board, row, col):
    return 0 <= row < 3 and 0 <= col < 3 and board[row][col] == ' '

# Function to get a valid player move
def get_player_move(board, player):
    while True:
        try:
            print(f"Player {player}, make your move (row and column):")
            row, col = map(int, input().split())
            if is_valid_move(board, row, col):
                return row, col
            else:
                print("Invalid move. Try again.")
        except ValueError:
            print("Invalid input. Please enter row and column as two integers.")

# Function to check if the game is won
def is_winner(board, player):
    for row in board:
        if all(cell == player for cell in row):
            return True
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

# Function to check if the game is a draw
def is_draw(board):
    return all(cell != ' ' for row in board for cell in row)

# Main game loop
def main():
    board = create_board()
    player = 'X'
    while True:
        display_board(board)
        row, col = get_player_move(board, player)
        board[row][col] = player
        if is_winner(board, player):
            display_board(board)
            print(f"Player {player} wins!")
            break
        if is_draw(board):
            display_board(board)
            print("It's a draw!")
            break
        player = 'O' if player == 'X' else 'X'

if __name__ == "__main__":
    main()
```
